```python
import sqlite3
```

```python
connection = sqlite3.connect('books.db')
```

```python
import pandas as pd
```

```python
pd.options.display.max_columns = 10
```

```python
pd.read_sql('SELECT * FROM authors', connection, index_col=['id'])
```

```python
pd.read_sql('SELECT * FROM titles', connection)
```

```python
pd.read_sql('SELECT * FROM author_ISBN', connection)
```

```python
query= """  SELECT title, edition, copyright
            FROM titles
            WHERE copyright > '2016' """
pd.read_sql(query, connection)
```

```python
query= """  SELECT id, first, last
            FROM authors
            WHERE last LIKE 'D%' """
pd.read_sql(query, connection, index_col=['id'])
```

```python
query=  """ SELECT id, first, last
            FROM authors
            WHERE first LIKE '_b%' """
pd.read_sql(query, connection, index_col = ['id'])
```

```python
pd.read_sql('SELECT title FROM titles ORDER BY title ASC', connection)
```

```python
query= """  SELECT id, first, last
            FROM authors
            ORDER BY last, first"""
pd.read_sql(query, connection, index_col=['id'])
```

```python
query= """  SELECT id, first, last
            FROM authors
            ORDER BY last DESC, first ASC"""
pd.read_sql(query, connection, index_col=['id'])
```

```python
query= """  SELECT isbn, title, edition, copyright
            FROM titles
            WHERE title LIKE '%How to Program'
            ORDER BY title"""
pd.read_sql(query, connection)
```
